```python
import numpy as np
import tensorflow as tf
import ast
```

```python
file_path = "/content/drive/My Drive/Colab Notebooks/interaktionDesign/game_results.txt"

with open(file_path, 'r') as file:
    content = file.read()
```

```python
lines = content.split('\n')
lines_tensor = tf.constant(lines)
shuffled_lines = tf.random.shuffle(lines_tensor)
print(shuffled_lines.shape)

all_inputs = []
all_labels = []
# Iteriere Ã¼ber jede Zeile und erstelle eine Trainingsdatei
for index, line in enumerate(shuffled_lines):
  line = line.numpy().decode('utf-8')

  input_str, label_str = line.split(';')
  input_list = np.array(ast.literal_eval(input_str))
  label = int(label_str)
  all_inputs.append(input_list)
  all_labels.append(label)

x_train = np.array(all_inputs[0:5000])  # Hier werden die Listen in numpy-Arrays umgewandelt
y_train = tf.keras.utils.to_categorical(np.array(all_labels[0:5000]), num_classes=3)

x_val = np.array(all_inputs[5000:])
y_val = tf.keras.utils.to_categorical(np.array(all_labels[5000:]), num_classes=3)

print(x_train.shape)
print(y_train.shape)
```

```python
print(y_train[2])
print(x_train[2])
```

```python
# Define the input layer
model = tf.keras.Sequential([
  tf.keras.layers.Input(shape=[9]),
  tf.keras.layers.Dense(units=9,activation="relu"),
  tf.keras.layers.Dense(units=3, activation="softmax")
])

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(0.01), metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=200, verbose=1, validation_data=(x_val, y_val) )
```
