```python
import numpy as np
import tensorflow as tf
```

```python
file_path = "/content/drive/My Drive/Colab Notebooks/interaktionDesign/wohnungen.txt"

with open(file_path, 'r') as file:
    content = file.read()
```

```python
lines = content.split('\n')
lines_tensor = tf.constant(lines)
shuffled_lines = tf.random.shuffle(lines_tensor)
print(shuffled_lines.shape)

all_inputs = []
all_labels = []
# Iteriere Ã¼ber jede Zeile und erstelle eine Trainingsdatei
for index, line in enumerate(shuffled_lines):
  line = line.numpy().decode('utf-8')

  input_str, label_str = line.split(';')
  input_list = list(map(float, input_str.split(',')))
  label = int(label_str)
  all_inputs.append(input_list)
  all_labels.append(label)

x_train = np.array(all_inputs[0:300])  # Hier werden die Listen in numpy-Arrays umgewandelt
y_train = tf.keras.utils.to_categorical(np.array(all_labels[0:300]), num_classes=2)

x_val = np.array(all_inputs[300:])
y_val = tf.keras.utils.to_categorical(np.array(all_labels[300:]), num_classes=2)

print(x_train.shape)
print(y_train.shape)
```

```python
for i in range(10):
  print(str(y_train[i]) + "," + str(x_train[i]))
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Input(shape=[3]),
  tf.keras.layers.Dense(units=4,activation="relu"),
  tf.keras.layers.Dense(units=2, activation="softmax")
])
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(0.01), metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=200, verbose=1, validation_data=(x_val, y_val) )
```

```python
print(model.predict([(1,2,250)]))
```
